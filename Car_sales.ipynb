#### Now we will calculate mean car price  for top 20 car brands.


```python
top_cars = autos['brand'].value_counts().head(20)

```

```python
autos['price_$'].describe()
```

Because on price column we overwrote value 0 with value 'Unknown' and now we need mean of all values in that column, below there is a function which pass rows with "Unknown" value.

```python
def mean_2(lista):
    avg = 0
    count = 0
    for row in lista:
        if row =='Unknown':
            pass
        else:
            avg += row
            count += 1
    final_avg = avg/count
    return final_avg
```

```python
top_cars_brands = top_cars.index


mean_price_brand = {}

for row in top_cars_brands:
    
        avg_price = mean_2(autos.loc[autos['brand'] == row, 'price_$'])
        mean_price_brand[row] = round(avg_price)

print(mean_price_brand)
```

Above there is a list with mean price brand. The most expensive ones are Mercedes(30,829), Citroen(45487) and

```python
top_6_brands_price = dict(itertools.islice(mean_price_brand.items(), 6))
```

```python
mean_mileage_brand = {}
for row in top_cars_brands:
    
    avg_mileage = mean_2(autos.loc[autos['brand'] == row, 'odometer_km'])
    mean_mileage_brand[row] = int(avg_mileage)
print(mean_mileage_brand)
```

```python
top_6_brands_mileage = dict(itertools.islice(mean_mileage_brand.items(), 6))
```

```python
series_brands_price = pd.Series(top_6_brands_price)
print(series_brands_price)
series_brands_mileage = pd.Series(top_6_brands_mileage)
print(series_brands_mileage)
```
